```python
#here the libraries
from sqlalchemy import create_engine
import pandas as pd
import getpass
password = getpass.getpass()
```

```python
connection_string = 'mysql+pymysql://root:' + password + '@localhost/sakila'
engine = create_engine(connection_string)
```

```python
#Create a Python connection with SQL database and 
#retrieve the results of the last two queries (also mentioned below) as dataframes
query = "select customer_id, count(rental_id) as rented_may from sakila.rentals_may \
group by customer_id order by rented_may desc"
data1 = pd.read_sql_query(query, engine)
data1 ##here the query for May rentals
```

```python
query = "select customer_id, count(rental_id) as rented_june from sakila.rentals_june \
group by customer_id order by rented_june desc"
data2 = pd.read_sql_query(query, engine)
data2 #here the query for June rentals
```

```python
#here to merge both df and create the "mergedf" dataframe
mergedf=data1.merge(data2, how='left', on='customer_id')
mergedf
```

```python
#here to check if there are more rentals on May or June
sum1 = mergedf['rented_may'].sum()
sum2 = mergedf['rented_june'].sum()
print (sum1)
print (sum2)
if sum1 > sum2:
    print("The amount of rented movies is higher on May")
elif sum1 < sum2:
    print("The amount of rented movies is higher on June")
else:
    print("Rented movies amount is the same for both months")
```
